```python
prompt = "A breathtaking sunset scene over a calm ocean, with the golden sun half-set on the horizon, casting vibrant shades of orange, pink, purple, and deep blue across the sky. Gentle waves reflect the warm hues of the sunset, creating a glowing pathway of light leading to the horizon. Silhouettes of distant seagulls fly across the sky, and a few palm trees sway gently in the foreground. The atmosphere feels peaceful, romantic, and serene, with a soft gradient of colors blending seamlessly. The lighting is warm and cinematic, emphasizing contrast between the glowing sky and darker silhouettes. Ultra-realistic, 8K resolution, high dynamic range, detailed reflections, volumetric lighting, and natural lens flare."
result = pipe(prompt).images[0]
display(result)
```
